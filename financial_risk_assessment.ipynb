```python
#importing neccessary imports
import numpy as np
import matplotlib.pyplot as plt     
import seaborn as sns   
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
```

```python
data = pd.read_csv('datasets/financial_risk_assessment.csv')
```

```python
plt.figure(figsize=(10, 6))
plt.hist(data['Age'], bins=30, color='skyblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['Credit Score'], data['Loan Amount'], c='blue', alpha=0.5)
plt.title('Credit Score vs Loan Amount')
plt.xlabel('Credit Score')
plt.ylabel('Loan Amount')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
data.boxplot(column='Income', by='Risk Rating', grid=False)
plt.title('Income Distribution by Risk Rating')
plt.suptitle('')  # Remove default title
plt.xlabel('Risk Rating')
plt.ylabel('Income')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Risk Rating', palette='Set2')
plt.title('Count of Risk Ratings')
plt.xlabel('Risk Rating')
plt.ylabel('Count')
plt.show()
```

```python
fig3 = px.scatter_3d(data, x='Age', y='Credit Score', z='Loan Amount', color='Risk Rating',
                     title="3D Scatter Plot: Age, Credit Score, and Loan Amount")
fig3.show()
```
